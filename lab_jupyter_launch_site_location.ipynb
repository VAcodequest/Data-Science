```python
# Task 2: Mark the success/failed launches for each site on the map
import folium
from folium.plugins import MarkerCluster

marker_cluster = MarkerCluster().add_to(site_map)

colors = {0: 'red', 1: 'green'}

for index, row in spacex_df.iterrows():
    location = [row['Lat'], row['Long']]
    marker_color = colors[row['class']]

    marker = folium.Marker(
        location=location,
        popup=f"Launch Site: {row['Launch Site']}<br>Outcome: {'Success' if row['class'] == 1 else 'Failure'}",
        icon=folium.Icon(color=marker_color)
    )
    marker.add_to(marker_cluster)

# Display the map
site_map
```

```python
# TASK 3: Calculate the distances between a launch site to its proximities
import folium
from geopy.distance import geodesic

reference_location = [28.5721, -80.6480]

map = folium.Map(location=reference_location, zoom_start=11, tiles='CartoDB positron')

folium.Marker(
    reference_location,
    popup="Kennedy Space Center",
    icon=folium.Icon(color="red", icon="info-sign")
).add_to(map)

def add_line(map, location1, location2, line_color="blue"):
    folium.PolyLine([location1, location2], color=line_color, weight=2.5, opacity=0.8).add_to(map)

for idx, row in spacex_df.iterrows():
    site_location = [row['Lat'], row['Long']]
    folium.Marker(
        site_location,
        popup=f"{row['Launch Site']}<br>Distance to KSC: {geodesic(site_location, reference_location).km:.2f} km",
        icon=folium.Icon(color='green' if row['class'] == 1 else 'red', icon='rocket')
    ).add_to(map)

    add_line(map, site_location, reference_location)

# Display the updated map
map


```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```
